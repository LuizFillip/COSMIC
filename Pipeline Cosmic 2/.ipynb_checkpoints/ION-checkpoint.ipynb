```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import datetime 
import os 
from glob import glob
import cartopy.crs as ccrs
import cartopy.feature as cfeature

class split_filename:
    
    def __init__(self, filename):
        
        self.filename = filename
        args = self.filename.split('.')
        
        self._leo = args[0]
        self._year = int(args[1])
        self._doy = int(args[2])
        self._hour = int(args[3])
        self._minute = int(args[4])
        
            
        self._date = datetime.date(self._year, 1, 1) + datetime.timedelta(self._doy - 1)
        
    
    @property
    def year(self):
        return self._year
    @property
    def doy(self):
        return self._doy
    @property
    def hour(self):
        return self._hour
    @property
    def minute(self):
        return self._minute
    @property
    def date(self):
        return self._date
args = split_filename(filename)
date = str(args.date).replace('-', '_')

```

```python
infile = "C:\\Users\\Jussiara Rodrigues\\Google Drive\\igaPrf.2020.260\\"
files = glob(infile + '/*.0001_nc')

ds = xr.open_dataset(files[0])

ds
```

```python
_, _, files = next(os.walk(infile))

lat = []
data = []
lon = []

for filename in files:

    if split_filename(filename).hour == 4:
        ds = xr.open_dataset(infile + filename)

        ds_mean = ds.mean()


        lat.append(ds_mean['GEO_lat'].values)
        lon.append(ds_mean['GEO_lon'].values)
        data.append(ds_mean['TEC_cal'].values)

```

```python
lat = np.array(lat)
lon = np.array(lon)
data = np.array(data)

fig, ax = plt.subplots(figsize = (12, 6), 
                       subplot_kw =dict(projection=ccrs.PlateCarree()))


plt.style.use('seaborn-talk')

ax.set_global()

states_provinces = cfeature.NaturalEarthFeature(
            category='cultural',
            name='admin_1_states_provinces_lines',
            scale='50m',
            facecolor='none')

ax.add_feature(cfeature.COASTLINE, edgecolor='black', lw = 1) 
ax.add_feature(cfeature.BORDERS, linestyle='-', edgecolor='black')
ax.add_feature(states_provinces, edgecolor='black')

ax.set(xticks = np.arange(-180, 210, 30), xlabel = 'longitude ()', ylabel = 'latitude ()',
       yticks = np.arange(-90, 110, 20), title = 'Radio Occultations - 2020-09-16 04:00 UT')

img = ax.scatter(lon, lat, c = data, s = 40, cmap = 'jet', transform = ccrs.PlateCarree())

cbar_ax = fig.add_axes([.9, 0.12, 0.02, 0.75]) 

fig.colorbar(img, cax=cbar_ax, label = 'TECU')

ax.scatter(-43.25, -13.11, s = 50, marker = 'X', color = 'k', 
           transform = ccrs.PlateCarree(), label = 'Bom Jesus da Lapa')

ax.legend()
plt.show()
```
